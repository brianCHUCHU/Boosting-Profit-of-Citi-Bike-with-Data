```python
import numpy as np

# 需要相加的欄位
sum_columns = ["bldgclass_A", "bldgclass_B", "bldgclass_C", "bldgclass_D", "bldgclass_E",
               "bldgclass_F", "bldgclass_G", "bldgclass_H", "bldgclass_I", "bldgclass_J",
               "bldgclass_K", "bldgclass_M", "bldgclass_N", "bldgclass_O", "bldgclass_P",
               "bldgclass_Q", "bldgclass_R", "bldgclass_S", "bldgclass_V", "bldgclass_W", "bldgclass_Y", "bldgclass_Z"]

df_rent_neighborhood["bldgclass_sum"] = df_rent_neighborhood[sum_columns].sum(axis=1)

# 不需要相加的欄位
mode_columns = ["NTA",
    "member_casual",
    "Total Population",
    "Population Density (per Sq. Mi.)",
    "Total Number of Households",
    "Street Mileage",
    "Estimated Number of Businesses",
    "Low-Income Housing (NYCHA)",
    "Number of Public Schools",
    "Street Furniture: Parking Pay Stations",
    "Street Furniture: Bus Stop Shelters",
    "Street Furniture: Bike Shelters",
    "Street Furniture: City Benches",
    "Empire City Subway Coverage (Percentage)",
    "Estimated Aerial Plant Coverage (Percentage)",
    "Estimated Underground Plant Coverage (Percentage)",
    "65+ population(%)",
    "65+ Below poverty(%)",
    "Hispanic/Latino(%)",
    "White(%)",
    "Black/ African American(%)",
    "Asian(%)",
    "Street Furniture: Parking Pay Stations", 
    "Street Furniture: Bus Stop Shelters", 
    "Street Furniture: Bike Shelters",
    "Street Furniture: City Benches", 
    ]

# 以NTA为基准进行groupby，计算每个group的数量和bldgclass_sum、mode_columns的平均值
grouped_df = df_rent_neighborhood.groupby('NTA').agg(
    group_count=('NTA', 'count'),
    bldgclass_sum=('bldgclass_sum', 'sum')
)

# 對mode_columns取平均
for col in mode_columns[1:]:
    grouped_df[col] = df_rent_neighborhood.groupby('NTA')[col].mean()

# 對sum_columns取平均
for col in sum_columns[1:]:
    grouped_df[col] = df_rent_neighborhood.groupby('NTA')[col].sum()

grouped_df.reset_index(inplace=True)
neighborhood_comparison = grouped_df
neighborhood_comparison
```

```python
# 不先移除相關係數相對高的變數，然後用forward selection來建立回歸模型
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso

# Assume df_rent_neighborhood is your DataFrame with all relevant variables including 'group_count' and 'NTA'
# Drop the constant column if it's already present
if 'const' in neighborhood_comparison:
    neighborhood_comparison = neighborhood_comparison.drop('const', axis=1)

# Separate the predictor variables (X) and the response variable (y)
X = neighborhood_comparison.drop(["group_count", "NTA"], axis=1)
y = neighborhood_comparison["group_count"]

# Standardize the predictor variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Forward Regression
significant_variables = []
remaining_variables = set(X.columns)

for i in range(len(X.columns)):
    best_pvalue = float('inf')
    best_variable = None
    
    for variable in remaining_variables:
        X_temp = sm.add_constant(X[significant_variables + [variable]])
        model = sm.OLS(y, X_temp)
        results = model.fit()
        
        if results.pvalues[variable] < best_pvalue:
            best_pvalue = results.pvalues[variable]
            best_variable = variable
    
    if best_pvalue < 0.05:  # Adjust the significance level as needed
        significant_variables.append(best_variable)
        remaining_variables.remove(best_variable)
    else:
        break

# Print significant variables
print("Significant Variables:", significant_variables)

# Use statsmodels to get the model summary
X_forward = sm.add_constant(X[significant_variables])
model_forward = sm.OLS(y, X_forward)
results_forward = model_forward.fit()
print(results_forward.summary())
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
import numpy as np
import statsmodels.api as sm

# Assuming X and y are defined based on 'neighborhood_comparison_filtered'
X = neighborhood_comparison_filtered.drop(columns=['group_count', 'NTA'])
y = neighborhood_comparison_filtered['group_count']

# Lasso Regression for Variable Selection
lasso_model = Lasso(alpha=1.0)  # You may need to tune the alpha parameter
lasso_model.fit(X, y)

# Print the coefficients
lasso_coefficients = lasso_model.coef_
print("Lasso Coefficients:", lasso_coefficients)

# Selected Features based on Lasso
selected_features = X.columns[np.abs(lasso_coefficients) > 0]
print("Selected Features (Lasso):", selected_features)

# Forward Selection using selected features
significant_variables_lasso = []
remaining_variables_lasso = set(selected_features)

for i in range(len(selected_features)):
    best_pvalue = float('inf')
    best_variable = None
    
    for variable in remaining_variables_lasso:
        X_temp = sm.add_constant(X[significant_variables_lasso + [variable]])
        model = sm.OLS(y, X_temp)
        results = model.fit()
        
        if results.pvalues[variable] < best_pvalue:
            best_pvalue = results.pvalues[variable]
            best_variable = variable
    
    if best_pvalue < 0.1:  # Adjust the significance level as needed
        significant_variables_lasso.append(best_variable)
        remaining_variables_lasso.remove(best_variable)
    else:
        break

# Print significant variables from Forward Regression using Lasso-selected features
print("Significant Variables (Forward Regression with Lasso-selected features):", significant_variables_lasso)

# Use statsmodels to get the model summary for forward regression
X_forward_lasso = sm.add_constant(X[significant_variables_lasso])
model_forward_lasso = sm.OLS(y, X_forward_lasso)
results_forward_lasso = model_forward_lasso.fit()
print(results_forward_lasso.summary())
```
